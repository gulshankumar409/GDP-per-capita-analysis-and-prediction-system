```python
GDP['GDP ($ per capita)'].fillna(value=GDP['GDP ($ per capita)'].mean(),inplace=True)
GDP['Literacy (%)'].fillna(GDP.groupby('Region')['Literacy (%)'].transform('median'), inplace= True)
GDP['Phones (per 1000)'].fillna(GDP.groupby('Region')['Phones (per 1000)'].transform('median'), inplace= True)
GDP['Birthrate'].fillna(GDP.groupby('Region')['Birthrate'].transform('mean'), inplace= True)
GDP['Deathrate'].fillna(GDP.groupby('Region')['Deathrate'].transform('mean'), inplace= True)
GDP['Agriculture'].fillna(value=GDP["Agriculture"].mean(),inplace=True)
GDP['Service'].fillna(value=GDP["Service"].mean(),inplace=True)
GDP['Net migration'].fillna(0, inplace=True)
GDP['Infant mortality (per 1000 births)'].fillna(0, inplace=True)
GDP['Arable (%)'].fillna(0, inplace=True)
GDP['Crops (%)'].fillna(0, inplace=True)
GDP['Other (%)'].fillna(0, inplace=True)
GDP['Climate'].fillna(0, inplace=True)
GDP['Industry'].fillna((1 - GDP['Agriculture'] - GDP['Service']), inplace= True)
```
